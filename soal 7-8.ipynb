```python
def isValidDate(year, month, day):
    # Define the maximum number of days for each month
    max_days = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

    # Check if the year is within a reasonable range
    if year < 1 or year > 999999:
        return False

    # Check if the month is within a valid range (1 to 12)
    if month < 1 or month > 12:
        return False

    # Check if the day is within a valid range for the given month
    if day < 1 or day > max_days[month]:
        return False

    # Special case: February in a leap year
    if month == 2 and year % 4 == 0 and (year % 100 != 0 or year % 400 == 0):
        if day > 29:
            return False

    return True

    # Test cases to verify the function
assert isValidDate(2001, 2, 29) == False
assert isValidDate(2029, 13, 1) == False
assert isValidDate(1000000, 1, 1) == False
assert isValidDate(2015, 4, 31) == False
```

```python
def commaFormat(number):
    # Check if the number is a float
    is_float = isinstance(number, float)

    # Convert the number to a string with proper comma formatting
    formatted_number = "{:,}".format(number)

    # If the number is a float, add back the fractional part
    if is_float:
        integer_part, fractional_part = formatted_number.split('.')
        formatted_number = f"{integer_part}.{fractional_part}"

    return formatted_number

# Test cases to verify the function
assert commaFormat(1) == '1'
assert commaFormat(10) == '10'
assert commaFormat(100) == '100'
assert commaFormat(1000) == '1,000'
assert commaFormat(10000) == '10,000'
assert commaFormat(100000) == '100,000'
assert commaFormat(1000000) == '1,000,000'
assert commaFormat(1234567890) == '1,234,567,890'
assert commaFormat(1000.123456) == '1,000.123456'
```
