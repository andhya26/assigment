```python
nums = [-1, 0, 1, -2, -1, 2, -4]
def sumList(nums):
  i = 0
  j = 0
  k = 0

  triplets = []
  for i in range(len(nums)):
    for j in range(i + 1, len(nums)):
      for k in range(j + 1, len(nums)):
        if nums[i] + nums[j] + nums[k] == 0:
          triplets.append([nums[i], nums[j],nums[k]])
```
